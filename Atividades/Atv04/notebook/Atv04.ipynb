# Exercício 01
<p style="text-align:justify;">
Crie um programa que armazene um vetor com as notas de dez alunos, calcule e imprima a média dessas notas e depois informe a quantidade de notas acima e abaixo da média calculada.
</p>

```java

import java.util.*;
// public class a {
    // public static void main(String[] args) {
        float notas[] = new float[10], soma = 0;
        int ba = 0, ci = 0;
        Scanner input = new Scanner(System.in);
        for (int i = 0; i < 10; i++) {
            System.out.println("Insira a nota:");
            notas[i] = input.nextFloat();
            soma += notas[i];
        }
        System.out.println("Media das notas:" + soma / 10);
        for (int i = 0; i < 10; i++) {
            if (notas[i] < soma / 10) {
                ba++;
            } else {
                ci++;
            }
        }
        System.out.println("Quantidade acima da media:" + ci + " Quantidade abaixo da media:" + ba);
        input.close();
    // }
// }
```
