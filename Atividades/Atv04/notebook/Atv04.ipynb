# Exercício 01
<p style="text-align:justify;">
Crie um programa que armazene um vetor com as notas de dez alunos, calcule e imprima a média dessas notas e depois informe a quantidade de notas acima e abaixo da média calculada.
</p>

```java


import java.util.Random;

// public class Notas {
    // public static void main(String[] args) {
        float notas[] = new float[10], soma = 0;
        int ba = 0, ci = 0;
       
        Random random = new Random();
        for (int i = 0; i < 10; i++) {
            notas[i] = random.nextFloat()*101;
            soma += notas[i];
        }
        System.out.println("Media das notas:" + soma / 10);
        for (int i = 0; i < 10; i++) {
            System.out.println(notas[i]);
            if (notas[i] < soma / 10) {
                ba++;
            } else {
                ci++;
            }
        }
        System.out.println("Quantidade acima da media:" + ci + " Quantidade abaixo da media:" + ba);
        
    // }
// }
```

# Exercício 02
<p style="text-align:justify;">
Construa uma classe em Java que seja capaz de efetuar a leitura de uma matriz de valores reais de tamanho 3 x 3, calculando seu determinante.
</p>

```java

import java.util.Random;

// public class Determinante {
    // public static void main(String[] args) {
        float mat[][] = new float[3][3];
        float diagPrin1, diagPrin2, diagPrin3, diagSec1, diagSec2, diagSec3;
        float det = 0;
        
        Random random = new Random();
        for (int i = 0; i < 3; i++) {
            for (int j = 0; j < 3; j++) {
                mat[i][j] = random.nextFloat()*101;
            }
        }
        for (int i = 0; i < 3; i++) {
            for (int j = 0; j < 3; j++) {
                if (j % 3 == 0) {
                    System.out.println();
                } else {
                    System.out.print("\t");
                }
                System.out.print(mat[i][j]);

            }
        }

        diagPrin1 = mat[0][0] * mat[1][1] * mat[2][2];
        diagPrin2 = mat[0][1] * mat[1][2] * mat[2][0];
        diagPrin3 = mat[0][2] * mat[1][0] * mat[2][1];

        diagSec1 = mat[2][0] * mat[1][1] * mat[0][2];
        diagSec2 = mat[2][1] * mat[1][2] * mat[0][0];
        diagSec3 = mat[2][2] * mat[1][0] * mat[0][1];

        det = (diagPrin1 + diagPrin2 + diagPrin3) - (diagSec1 + diagSec2 + diagSec3);
        System.out.println();
        System.out.println("Determinante:" + det);
        

    // }
// }
```

# Exercício 03
<p style="text-align:justify;">
Faça uma classe contendo dois arrays do tipo inteiro, um chamado par e outro chamado ímpar. Cada array deve conter 10 elementos. A seguir, faça um laço de repetição com 10 iterações e que contenha internamente a geração randômica de números entre 1 e 20. Se o valor gerado for par, armazene no array chamado par, caso contrário, no array chamado impar. Ao final, apresente o conteúdo de cada um dos arrays.
</p>

```java

import java.util.Random;

// public class ParImpar {
    // public static void main(String[] args) {
        int par[] = new int[10], impar[] = new int[10];
        int x;
        Random random = new Random();
        for (int i = 0; i < 10; i++) {
            x = random.nextInt(21);
            if (x % 2 == 0) {
                par[i] = x;
            } else {
                impar[i] = x;
            }

        }
        System.out.println("Par");
        for (int i = 0; i < 10; i++) {
            if (par[i] != 0) {
                System.out.println(par[i]);
            }
        }
        System.out.println("Impar");
        for (int i = 0; i < 10; i++) {
            if (impar[i] != 0) {
                System.out.println(impar[i]);
            }
        }

    // }
// }
```
